# FEATURE ENGINEERING

```python
#feature selection. dropping features not required.
avocado.drop(['Date','region'],axis = 1, inplace = True)
```

```python
#checking shape after dropping of columns
avocado.shape
```

# ENCODING

```python
#converting categorical columns into integer type column by using label encorder
le = preprocessing.LabelEncoder() #creating an instance
```

```python
avocado['type'] = le.fit_transform(avocado['type']) #transforming the data
```

# HANDLING SKEWNESS AND OUTLIERS

```python
#handling skewness
#values having z score more than 3 are removed 
z = np.abs(zscore(avocado))
avo= avocado[(z<3).all(axis=1)]
avo
```

```python
#revised and reduced skewdata 
avo.skew()
```

# checking for duplicate data

```python
#no duplicate values are present
duplicate = avo[avo.duplicated(['AveragePrice', 'Total Volume', '4046', '4225', '4770', 'Total Bags',
       'Small Bags', 'Large Bags', 'XLarge Bags', 'type', 'year'])]
duplicate
```

# SPLITTING DATA

```python
#splitting the label column
#label column is average price
x = avo.iloc[:,1:14]
x.head()
```

```python
#t part contains only dpendent variable which is average price
y = avo.iloc[:,0]
y.head()
```

# TRAINING DATA

```python
#training x and y 
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=47)
```

```python
#scaling of data and checking if it fits and transforming split train data
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_train_sc
```

```python
#scaling and transforming split test data
x_test_sc = sc.transform(x_test)
x_test_sc
```

# LINEAR REGRESSION

```python
#linear regression - model1
lm = LinearRegression() #creating an instance
```

```python
lm.fit(x_train_sc,y_train) #fitting linear regression model into trained data
lm.coef_ #coefficient of x
```

```python
lm.intercept_    #intercept value of the regreesion model
```

```python
lm.score(x_train_sc,y_train)  #as per the score, our model works 41% well
```

```python
#predict the value
predy = lm.predict(x_test_sc)
print("predicted result price:",predy)
print('actual price:',y_test)
```

```python
#determing error values, cv score and r2 score
print('r2 score =',r2_score(y_test,predy))
print('mean absolute error =',mean_absolute_error(y_test,predy))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predy)))
print('cv score=',cross_val_score(lm, x, y, cv=10))
```

# RIDGE REGRESSION

```python
#Ridge regression
rd = Ridge(alpha = 0.0001) #creating an instance
rd.fit(x_train_sc,y_train)  #fitting the model into trained data
rd.score(x_train_sc,y_train)  #as per the score, our model works 41% well
```

```python
#predict the value
predy_rd = rd.predict(x_test_sc)
print("predicted result price:",predy_rd)
print('actual price:',y_test)
```

```python
#determing error values, cv score and r2 score
print('r2 score =',r2_score(y_test,predy_rd))
print('mean absolute error =',mean_absolute_error(y_test,predy_rd))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predy_rd)))
print('cv score=',cross_val_score(rd, x, y, cv=10))
```

# HYPERPARAMETER TUNING

```python
#as random forest classifier is the best model. thus we are proceeding with its tuning
#we are using grid search cv
RFC = RandomForestRegressor() #creating an instance
n_estimators = [0.1,10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grid = dict(n_estimators = n_estimators)
gridf = GridSearchCV(estimator = RFC, param_grid = grid, n_jobs=-1, verbose = 2, cv = 5)
gridf.fit(x_train_sc,y_train) #fitting the model into trained data
```

```python
gridf.best_params_
```

```python
gridf.best_score_
```

```python
gridf.score(x_train_sc,y_train)   #as per the score, our model works 97% well
```

```python
#determing error values, cv score and r2 score
print("predicted result price:",pred_g)
print('actual price:',y_test)
print('r2 score =',r2_score(y_test,pred_g))
print('mean absolute error =',mean_absolute_error(y_test,pred_g))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,pred_g)))
print('cv score=',cross_val_score(RFC, x, y, cv=10))
```
