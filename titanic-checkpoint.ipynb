```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import klib
from scipy.stats import skew
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import confusion_matrix
import joblib
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#handling mssing values
df['Age'].fillna(df['Age'].mean(),inplace = True)
df['Cabin'].fillna(df['Cabin'].mode(),inplace = True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace = True)
```

```python
#feature selection. dropping features not required.
df.drop(['Cabin','Ticket','Name'],axis = 1, inplace = True)
```

# ENCODING OF CATEGORICAL COLUMNS

```python
#among 5 categorical columns, 3 were unnecessary and was dropped earlier.
#remaing 2, i'e sex and embarked are needed to be encoded
df['Embarked'].value_counts() #checking value count of embarked
```

```python
#assigning numerical values to the categorical values of columns in the dictionary
data = {'Sex':{'male':1,'female':0},'Embarked':{'S':1,'C':2,'Q':3}}
```

```python
#replacing categorical data with assigned numerical data
data1 = df.replace(data)
data1.head()
```

# CHECKING MULTICOLLINEARITY

```python
#checking correlation.
#none of the values are positively highly correlated.
#thus we can rule out the possibility of multicollinearity
correlation = data1.corr()
correlation
```

```python
#expressing collinearity in a graph format
plt.figure(figsize=(15,8))
sn.heatmap(data1.corr(), annot=True, linewidths=0.5)
```

# HANDLING SKEWNESS

```python
#we observed that 4 features are highly skewed when compared to others. so we are reducing skewness using square root process. 
#skewing fare
data1['Fare'] = np.sqrt(data1['Fare'])
skew(data1['Fare'])
```

```python
#skewing parch
data1['Parch']= np.sqrt(data1['Parch'])
skew(data1['Parch'])
```

```python
#skewing embarked
data1['Embarked']= np.sqrt(data1['Embarked'])
skew(data1['Embarked'])
```

```python
#skewing sibsp
data1['SibSp']= np.sqrt(data1['SibSp'])
skew(data1['SibSp'])
```

```python
#now checking the skewness of the data after reducing skewness. checking to see if it worked
klib.dist_plot(data1)
```

```python
#skewness is reduced as seen in the graphs
```

# splitting data

```python
#partitioning
#splitting the train data into train and test
#dividing data>>>
x = data1.drop(columns =['Survived','PassengerId'], axis = 1)
y = data1['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

# scaling of data

```python
#scaling of data and checking if it fits and transforming split train data
sc = StandardScaler()
x_trainstd = sc.fit_transform(x_train)
x_trainstd
```

```python
#scaling and transforming split test data
x_teststd = sc.transform(x_test)
x_teststd
```

# MODEL BULIDING

# RANDOM FOREST CLASSIFIER

```python
model_clf = RandomForestClassifier(n_estimators=100)
model_clf.fit(x_trainstd,y_train)
```

```python
clf_pred=model_clf.predict(x_teststd)
```

```python
print('accuracy=', metrics.accuracy_score(y_test,clf_pred))
#accuracy of this model is 81%
```

# DECISION TREE CLASSIFIER

```python
model_tree = DecisionTreeClassifier()
model_tree.fit(x_trainstd,y_train)
```

```python
tree_pred=model_tree.predict(x_teststd)
```

```python
print('accuracy=', metrics.accuracy_score(y_test,tree_pred))
#accuracy of this model is 75%
```

# KNEIGHBORS CLASSIFIER

```python
model_KNN = KNeighborsClassifier(n_neighbors = 5)
model_KNN.fit(x_trainstd,y_train)
```

```python
KNN_pred=model_KNN.predict(x_teststd)
```

```python
print('accuracy=', metrics.accuracy_score(y_test,KNN_pred))
#accuracy of this model is 83%
```

# LOGISTIC REGRESSION

```python
model_LOR = LogisticRegression(solver='liblinear',random_state=0)
model_LOR.fit(x_trainstd,y_train)
```

```python
LOR_pred=model_LOR.predict(x_teststd)
```

```python
print('accuracy=', metrics.accuracy_score(y_test,LOR_pred))
#accuracy of this model is 79%
```

# HYPERPARAMETER TUNING

```python
#as random forest classifier has highest accuracy. thus we are proceeding with its tuning
#we are using grid search cv
model_forest = RandomForestClassifier()
n_estimators = [10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grids = dict(n_estimators = n_estimators)
```

```python
gridforest = GridSearchCV(estimator = model_forest, param_grid = grids, n_jobs=-1, verbose = 2, cv = 3)
```

```python
gridforest.fit(x_trainstd,y_train)
```

```python
gridforest.best_params_
```

```python
gridforest.best_score_
```

```python
newgrid = gridforest.predict(x_teststd)
newgrid
```

```python
print('accuracy=', metrics.accuracy_score(y_test,newgrid))
#final score is 82%. hence we will deploy it 
```

# plotting the predicted data using confusion matrix

```python
comat = confusion_matrix(y_test,newgrid)
col = ['Survived','Not Survived']
row = ['Predicted Survived','Predicted Not Survived']
comat_frame = pd.DataFrame(comat,index = row, columns = col)
```

```python
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, yscore)
```
