```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/happiness_score_dataset.csv')
df.head()
```

```python
df.to_csv("happiness_report_dataset.csv")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df.shape
```

```python
drop_col = ['Country','Region','Happiness Rank','Happiness Score','Standard Error','Dystopia Residual']
features = df.drop(drop_col,axis = 1)
target = df['Happiness Score']
features
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_data = features.to_numpy()
y_data = target.to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data,test_size = 0.2)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
coefficients = list()
coefficients = reg.coef_
bias = reg.intercept_
print(coefficients)
print(bias)
```

```python
HS_predicted = reg.predict(X_data)
```

```python
HS_predicted
```

```python
HS_df_predicted = pd.DataFrame(HS_predicted)
```

```python
df['Happiness Score (predicted)'] = HS_df_predicted
```
