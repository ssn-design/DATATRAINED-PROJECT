```python
#using label encoding to encode the categorical values
le=LabelEncoder() #creating an instance
#fitting and transforming the categorical columns.
ins['policy_csl']=le.fit_transform(ins['policy_csl'])
ins['policy_state']=le.fit_transform(ins['policy_state'])
ins['policy_bind_date']=le.fit_transform(ins['policy_bind_date'])
ins['insured_sex']=le.fit_transform(ins['insured_sex'])
ins['insured_education_level']=le.fit_transform(ins['insured_education_level'])
ins['insured_occupation']=le.fit_transform(ins['insured_occupation'])
ins['insured_hobbies']=le.fit_transform(ins['insured_hobbies'])
ins['insured_relationship']=le.fit_transform(ins['insured_relationship'])
ins['incident_date']=le.fit_transform(ins['incident_date'])
ins['incident_location']=le.fit_transform(ins['incident_location'])
ins['incident_type']=le.fit_transform(ins['incident_type'])
ins['collision_type']=le.fit_transform(ins['collision_type'])
ins['incident_severity']=le.fit_transform(ins['incident_severity'])
ins['authorities_contacted']=le.fit_transform(ins['authorities_contacted'])
ins['incident_state']=le.fit_transform(ins['incident_state'])
ins['incident_city']=le.fit_transform(ins['incident_city'])
ins['property_damage']=le.fit_transform(ins['property_damage'])
ins['police_report_available']=le.fit_transform(ins['police_report_available'])
ins['auto_make']=le.fit_transform(ins['auto_make'])
ins['auto_model']=le.fit_transform(ins['auto_model'])
ins['fraud_reported']=le.fit_transform(ins['fraud_reported'])
```

# SPLITTING DATA

```python
x = ins.drop(columns =['fraud_reported'], axis = 1)
y = ins['fraud_reported']
```

# PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
x=pca.fit_transform(x)
```

# TRAINING DATA

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.3)
```

# SCALING DATA

```python
sc=StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

# MODEL BUILDING - 1.RANDOM FOREST CLASSIFIER

```python
rfc = RandomForestClassifier(n_estimators=100)#creating an instance
rfc.fit(x_train_sc,y_train)  #fitting random forest model into trained data
```

```python
y_rfc_pred=rfc.predict(x_test_sc) #predicting testing data
print('accuracy=', metrics.accuracy_score(y_test,y_rfc_pred))
print('cv score=',cross_val_score(rfc, x, y, cv=5))
# our models is working well for 84%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat_rfc = confusion_matrix(y_test,y_rfc_pred)
col = ['fraud not reported','fraud reported']
row = ['predicted fraud reported','Predicted fraud not reported']
comat_frame = pd.DataFrame(comat_rfc,index = row, columns = col)
plt.figure.figsize = (10,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore1 = rfc.predict_proba(x_test_sc)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, yscore1)
print('roc_auc_score for random forest classifier is ', roc_auc_score(y_test, yscore1))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('random forest - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# MODEL BUILDING- 2.DECISION TREE CLASSIFIER

```python
dtree = DecisionTreeClassifier() #creating an instance
dtree.fit(x_train_sc,y_train)  #fitting random forest model into trained data
ytree_pred=dtree.predict(x_test_sc)  #predicting testing data
print('accuracy=', metrics.accuracy_score(y_test,ytree_pred))
print('cv score=',cross_val_score(dtree, x, y, cv=5))
#accuracy of this model is 78%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat_tr = confusion_matrix(y_test,ytree_pred)
col = ['fraud not reported','fraud reported']
row = ['Predicted fraud reported','Predicted fraud not reported']
comat_frame = pd.DataFrame(comat_tr,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('decision tree - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# HYPERPARAMETER TUNING

```python
#as logistic regression has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
lor_h = LogisticRegression()
solvers = ['newton-cg','lbfgs','liblinear']
penalty = ['l1','l2']
c_values = [100,10,1.0,0.1,0.01,0.001,0.0001]
grid = dict(solver = solvers, penalty = penalty, C=c_values)
cv = RepeatedStratifiedKFold(n_splits =10,n_repeats = 5,random_state=42)
grid1 = GridSearchCV(estimator = lor_h, param_grid = grid, n_jobs=-1, verbose = True, cv = cv, scoring = 'accuracy', error_score = 0)
grid1.fit(x_train_sc,y_train) #fitting into trained data
```

```python
grid1.best_params_ #selecting best parameter
```

```python
y_grid = grid1.predict(x_test_sc) #predicting test data 
y_grid
```

```python
print('accuracy=', metrics.accuracy_score(y_test,y_grid))
print('cv score=',cross_val_score(lor_h, x, y, cv=5))
# our models is working well for 85%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat_LR_h = confusion_matrix(y_test,y_grid)
col = ['fraud not reported','fraud reported']
row = ['Predicted fraud reported','Predicted fraud not reported']
comat_frame = pd.DataFrame(comat_LR_h,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore5 = grid1.predict_proba(x_test_sc)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, yscore5)
print('roc_auc_score for hypertuned LOGISTIC REGRESSION classifier is ', roc_auc_score(y_test, yscore5))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('LOGISTIC REGRESSION - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
