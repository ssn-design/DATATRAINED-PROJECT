# Feature engineering

```python
#using label encoding to encode the categorical values
le=LabelEncoder() #creating an instance
#fitting and transforming the categorical columns.
la['Gender']=le.fit_transform(la['Gender'])
la['Married']=le.fit_transform(la['Married'])
la['Dependents']=le.fit_transform(la['Dependents'])
la['Education']=le.fit_transform(la['Education'])
la['Loan_ID']=le.fit_transform(la['Loan_ID'])
la['Self_Employed']=le.fit_transform(la['Self_Employed'])
la['Property_Area']=le.fit_transform(la['Property_Area'])
la['Loan_Status']=le.fit_transform(la['Loan_Status'])
```

# multi-collinearity

```python
#separating target column
a = la.drop(columns =['Loan_Status'], axis = 1, inplace=False)
```

```python
#creating a function to find vif factor to check multi-collinearity
def calvif():
    vif=pd.DataFrame()
    vif["VIF Factor"]=[variance_inflation_factor(a.values,i) for i in range(a.shape[1])]
    vif["Features"]=a.columns
    print(vif)
```

```python
calvif() #calling the function
```

```python
#heatmap representing correlation table.
plt.figure(figsize=(15,7))
sn.heatmap(la.corr(),annot=True, linewidths=0.5)
```

```python
#dropping high variance columns with low contribution
#dropping Loan_Amount_Term as it has high variance with low contribution.
la.drop(columns=['Loan_Amount_Term'],axis =1, inplace =True)
```

# skewness

```python
#checking skewness
la.skew()
```

```python
#reducing skewness by using z score. keeping values who have z score more than threshold 3
z = np.abs(zscore(la))
la1= la[(z<3).all(axis=1)]
la1.skew()
```

# SPLITTING TRAIN & TEST DATA

```python
#splitting data to train and test for model building
x = la1.drop(columns =['Loan_Status'], axis = 1)
y = la1['Loan_Status']
```

# TRAINING DATA

```python
#training and testing data x and y
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.3)
```

# STANDARDIZING DATA

```python
#normalizing data 
sc=StandardScaler() #creating an instance
x_train_s = sc.fit_transform(x_train)
x_test_s = sc.transform(x_test)
```

# MODEL BUILDING - 1. KNN NEIGHBORS CLASSIFIER

```python
KN = KNeighborsClassifier(n_neighbors = 5) #creating an instance
KN.fit(x_train_s,y_train)   #fitting knn model into trained data
Y_KN_pred=KN.predict(x_test_s)   #predicting testing data
print('accuracy=', metrics.accuracy_score(y_test,Y_KN_pred))
print('cv score=',cross_val_score(KN, x, y, cv=5))
#accuracy of this model is 79%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - loan status approved, no- loan status not approved of actual and predicted y value
comat_1 = confusion_matrix(y_test,Y_KN_pred)
col = ['LOAN NOT APPROVED','LOAN APPROVED']
row = ['Predicted LOAN APPROVED','Predicted LOAN NOT APPROVED']
comat_frame = pd.DataFrame(comat_1,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore1 = KN.predict_proba(x_test_s)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, yscore1)
print('roc_auc_score for KN NEIGHBORS classifier is ', roc_auc_score(y_test, yscore1))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('KNN - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 2.DECISION TREE CLASSIFIER

```python
dtr = DecisionTreeClassifier() #creating an instance
dtr.fit(x_train_s,y_train)  #fitting decision tree model into trained data
ytr_pred=dtr.predict(x_test_s)  #predicting testing data
print('accuracy=', metrics.accuracy_score(y_test,ytr_pred))
print('cv score=',cross_val_score(dtr, x, y, cv=5))
#accuracy of this model is 70.6%
```

```python
# creating the confusion matrix which is a graphical 
#representation of data of yes - loan approved, no-loan not approved of actual and predicted y value
comat_2= confusion_matrix(y_test,ytr_pred)
col = ['LOAN NOT APPROVED','LOAN APPROVED']
row = ['Predicted LOAN APPROVED','Predicted LOAN NOT APPROVED']
comat_frame = pd.DataFrame(comat_2,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore2 = dtr.predict_proba(x_test_s)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, yscore2)
print('roc_auc_score for decision tree classifier is ', roc_auc_score(y_test, yscore2))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('decision tree - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# HYPERTUNING

```python
#as random forest classifier is the best model and has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
RFC = RandomForestClassifier() #creating an instance
n_estimators = [0.1,10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grid = dict(n_estimators = n_estimators)
grid_f = GridSearchCV(estimator = RFC, param_grid = grid, n_jobs=-1, verbose = 2, cv = 5)
grid_f.fit(x_train_s,y_train) #fitting the model into trained data
```
