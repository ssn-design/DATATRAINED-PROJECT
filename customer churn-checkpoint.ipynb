# check multi-collinearity

```python
#dropping the target column to find VIF
x1 = cx.drop(columns =['Churn'], axis = 1, inplace=False)
```

```python
#creating a fuction to find vif factor to check multi-collinearity
def vif_cal():
    viff=pd.DataFrame()
    viff["VIF FACTOR"]=[variance_inflation_factor(x1.values,i) for i in range(x1.shape[1])]
    viff["Features"]=x1.columns
    print(viff)
```

```python
vif_cal() #calling the function
```

```python
#now checking it with heatmap of correlation to find out high variance creating columns.
plt.figure(figsize=(25,15))
sn.heatmap(cx.corr(),annot=True, linewidths=0.5)
```

# splitting and training dataset

```python
X = cx.drop(columns =['Churn'], axis = 1)
Y = cx['Churn']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=42, test_size=0.3)
```

# STANDARDIZING DATA

```python
#scaling of data and checking if it fits and transforming split train and test data
sc = StandardScaler()
X_train1 = sc.fit_transform(X_train)
X_test1 = sc.transform(X_test)
```

# MODEL BUILDING - CLASSIFICATION

# 1.DECISION TREE CLASSIFIER

```python
dtr = DecisionTreeClassifier() #creating an instance
dtr.fit(X_train1,Y_train)  #fitting random forest model into trained data
ytr_pred=dtr.predict(X_test1)  #predicting testing data
print('accuracy=', metrics.accuracy_score(Y_test,ytr_pred))
print('cv score=',cross_val_score(dtr, X, Y, cv=5))
#accuracy of this model is 71%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat1 = confusion_matrix(Y_test,ytr_pred)
col = ['CUSTOMER STAY','CUSTOMER LEFT']
row = ['PredicteD CUSTOMER LEFT','Predicted CUSTOMER STAY']
comat_frame = pd.DataFrame(comat1,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscoreTR = dtr.predict_proba(X_test1)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(Y_test, yscoreTR)
print('roc_auc_score for decision tree classifier is ', roc_auc_score(Y_test, yscoreTR))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('decision tree - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 2 - LOGISTIC REGRESSION

```python
LR = LogisticRegression(solver='liblinear',random_state=0) #creating an instance
LR.fit(X_train1,Y_train)   #fitting random forest model into trained data
YL_pred=LR.predict(X_test1)   #predicting testing data
print('accuracy=', metrics.accuracy_score(Y_test,YL_pred))
print('cv score=',cross_val_score(LR, X, Y, cv=5))
#accuracy of this model is 79%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat_LR = confusion_matrix(Y_test,YL_pred)
col = ['CUSTOMER LEFT','CUSTOMER STAY']
row = ['Predicted CUSTOMER STAY','Predicted CUSTOMER LEFT']
comat_frame = pd.DataFrame(comat_LR,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore_LR = LR.predict_proba(X_test1)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(Y_test, yscore_LR)
print('roc_auc_score for LOGISTIC REGRESSION classifier is ', roc_auc_score(Y_test, yscore_LR))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('LOGISTIC REGRESSION - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# HYPERPARAMETER TUNING

```python
#as logistic regression has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
lor1 = LogisticRegression()
solvers = ['newton-cg','lbfgs','liblinear']
penalty = ['l1','l2']
c_values = [100,10,1.0,0.1,0.01,0.001,0.0001]
grid = dict(solver = solvers, penalty = penalty, C=c_values)
cv = RepeatedStratifiedKFold(n_splits =10,n_repeats = 5,random_state=42)
gridS = GridSearchCV(estimator = lor1, param_grid = grid, n_jobs=-1, verbose = True, cv = cv, scoring = 'accuracy', error_score = 0)
gridS.fit(X_train1,Y_train) #fitting into trained data
```

```python
gridS.best_params_ #selecting best parameter
```

```python
gridy = gridS.predict(X_test1) #predicting test data 
gridy
```

```python
print('accuracy=', metrics.accuracy_score(Y_test,gridy))
print('cv score=',cross_val_score(lor1, X, Y, cv=5))
# our models is working well for 79.33%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat_lor1 = confusion_matrix(Y_test,gridy)
col = ['CUSTOMER LEFT','CUSTOMER STAY']
row = ['Predicted CUSTOMER STAY','Predicted CUSTOMER LEFT']
comat_frame = pd.DataFrame(comat_lor1,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore_lor1 = gridS.predict_proba(X_test1)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(Y_test, yscore_lor1)
print('roc_auc_score for logistic regression is ', roc_auc_score(Y_test, yscore_lor1))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('hypertuned logistic regression - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
