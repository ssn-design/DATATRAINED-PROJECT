# checking multicollinearity

```python
#separating the target variables from dataset
t2=['Next_Tmax','Next_Tmin']
y1=temp[t2]
x1=temp.drop(t2,axis=1)
```

```python
#creating a function to find vif factor to check multi-collinearity
def vif_cal():
    viff=pd.DataFrame()
    viff["VIF FACTOR"]=[variance_inflation_factor(x1.values,i) for i in range(x1.shape[1])]
    viff["Features"]=x1.columns
    print(viff)
```

```python
#calling the function
vif_cal()
```

we see that:
    
    1. lat and lon
    
    2. LDAPS_Tmax_lapse and LDAPS_Tmin_lapse
    
    3. LDAPS_RHmax and  Present_Tmin
    
    4. Present_Tmax and Solar radiation
    
    5. LDAPS_RHmin and LDAPS_WS
    
pairs have high VIF value. thus we will crosscheck with correlation table and drop out 1 column from each of these pair who contribute less to the target variables.
lon,LDAPS Tmin lapse, LDAPS RHmax, solar radiation,LDAPS RHmin contribute less when compared to the dataset. thus dropping those columns

```python
#dropping columns creating high variance with less contribution to reduce multi collinearity
temp.drop(columns=['lon','LDAPS_Tmin_lapse','LDAPS_RHmax','Solar radiation','LDAPS_RHmin'], axis = 1,inplace= True)
```

```python
#checkinh shape after dropping
temp.shape
```

# splitting of dataset

```python
#separating the target variables from dataset into x and y. y is the target variable. x does not contain target variable
target=['Next_Tmax','Next_Tmin']
y=temp[target]
x=temp.drop(target,axis=1)
```

# PCA

```python
#to reduce the no. of columns from 21 to 10
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
x=pca.fit_transform(x)
```

# training data

```python
#training data
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

# SCALING OF DATA

```python
#scaling of data and checking if it fits and transforming split train and test data
sc = StandardScaler()
x_trains = sc.fit_transform(x_train)
x_tests = sc.transform(x_test)
```

# MODEL BUILDING: 1.LINEAR REGRESSION

```python
lin = LinearRegression() #creating an instance
lin.fit(x_trains,y_train) #fitting linear regression model into trained data
lin.coef_ #coefficient of x
```

```python
lin.intercept_  #intercept value of the regreesion model
```

```python
lin.score(x_trains,y_train) #as per the score, our model works 71% well
```

```python
#predict the value
predicted_y_lin = lin.predict(x_tests)
print("predicted results:",predicted_y_lin)
print('actual result:',y_test)
print('r2 score =',r2_score(y_test,predicted_y_lin))
print('mean absolute error =',mean_absolute_error(y_test,predicted_y_lin))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predicted_y_lin)))
print('cv score=',cross_val_score(lin, x, y, cv=3))
```

# KNEIGHBOURS

```python
KNEIGHBOURS = KNeighborsRegressor(n_neighbors=7) #creating an instance
KNeighborsRegressor(algorithm='auto', leaf_size=30, metric='minkowski',
          metric_params=None, n_jobs=1, n_neighbors=7, p=2,
          weights='uniform')
KNEIGHBOURS.fit(x_trains,y_train) #fitting kneighbors into x train and y train values
KNEIGHBOURS.score(x_trains,y_train)  #as per the score, our model works 85% well
```

```python
#predict the value
predicted_y_kn= KNEIGHBOURS.predict(x_tests)
print("predicted results:",predicted_y_kn)
print('actual result:',y_test)
print('r2 score =',r2_score(y_test,predicted_y_kn))
print('mean absolute error =',mean_absolute_error(y_test,predicted_y_kn))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predicted_y_kn)))
print('cv score=',cross_val_score(KNEIGHBOURS, x, y, cv=5))
```

# RANDOM FOREST REGRESSOR

```python
RFR = RandomForestRegressor() #creating an instance
RFR.fit(x_trains,y_train) #fitting random forest regressor into x train and y train values
RFR.score(x_trains,y_train)  #as per the score, our model works 97% well
```

```python
#predict the value, r2 and error scores
y_RFR = RFR.predict(x_tests)
print("predicted results:",y_RFR)
print('actual results:',y_test)
print('r2 score =',r2_score(y_test,y_RFR))
print('mean absolute error =',mean_absolute_error(y_test,y_RFR))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,y_RFR)))
print('cv score=',cross_val_score(RFR, x, y, cv=5))
```

# ELASTIC NET REGRESSION

```python
en = ElasticNet(alpha = 0.0001) #creating an instance
en.fit(x_trains,y_train)   #fitting into trained data
en.score(x_trains,y_train)  #as per the score, our model works 71% well
```

```python
#predict the value, r2 and error scores
y_ENT = en.predict(x_tests)
print("predicted results:",y_ENT)
print('actual results:',y_test)
print('r2 score =',r2_score(y_test,y_ENT))
print('mean absolute error =',mean_absolute_error(y_test,y_ENT))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,y_ENT)))
print('cv score=',cross_val_score(en, x, y, cv=5))
```

# HYPERPARAMETER TUNING

```python
#as random forest regressor is the best model and has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
RDM = RandomForestRegressor() #creating an instance
n_estimators = [0.1,10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grid = dict(n_estimators = n_estimators)
grid1 = GridSearchCV(estimator = RDM, param_grid = grid, n_jobs=-1, verbose = 2, cv = 7)
grid1.fit(x_trains,y_train) #fitting the model into trained data
```

```python
grid1.best_params_ #getting best parameter
```

```python
grid1.best_score_
```

```python
grid1.score(x_trains,y_train)   #as per the score, our model works 97.9% well
```

```python
#predict the value
y_gridp = grid1.predict(x_tests)
y_gridp
```

```python
#predict the value, r2 and error scores
print("predicted results:",y_gridp)
print('actual results:',y_test)
print('r2 score =',r2_score(y_test,y_gridp))
print('mean absolute error =',mean_absolute_error(y_test,y_gridp))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,y_gridp)))
print('cv score=',cross_val_score(grid1, x, y, cv=3))
```
