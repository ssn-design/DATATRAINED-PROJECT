# feature engineering

```python
census1['Education']=census1['Education'].replace(['HS-grad','11th','10th','7th-8th','9th','12th','5th-6th','1st-4th','Preschool'],'School')
census1['Education']=census1['Education'].replace(['Some-college','Assoc-acdm','Prof-school','Assoc-voc','Bachelors'],'college')
census1['Education']=census1['Education'].replace(['Masters','Doctorate'],'postgrad')
```

```python
census1.Relationship=census1.Relationship.replace(['Wife','Husband','Own-child','Other-relative'],'family')
census1.Relationship=census1.Relationship.replace('Not-in-family','not_family')
```

```python
census1['Marital_status']=census1['Marital_status'].replace(['Married-civ-spouse','Married-AF-spouse'],'married')
census1['Marital_status']=census1['Marital_status'].replace('Never-married','unmarried')
census1['Marital_status']=census1['Marital_status'].replace(['Divorced','Separated','Widowed','Married-spouse-absent'],'married_but_single')
```

```python
census1['Native_country']=census1['Native_country'].replace(['Mexico','Honduras','Canada','Guatemala','Columbia','Nicaragua','Peru','Ecuador'],'northamerica')
census1['Native_country']=census1['Native_country'].replace(['Philippines','India','China','Vietnam','Japan','Taiwan','Iran','Hong','Cambodia','Laos','Thailand'],'asia')
census1['Native_country']=census1['Native_country'].replace(['Germany','Hungary','England','Italy','Poland','Portugal','France','Greece','Ireland','Yugoslavia','Holand-Netherlands'],'europe')
census1['Native_country']=census1['Native_country'].replace(['Puerto-Rico','El-Salvador','Haiti','Cuba','Jamaica','Dominican-Republic','Trinadad&Tobago'],'carribean')
census1['Native_country']=census1['Native_country'].replace('South','other')
census1['Native_country']=census1['Native_country'].replace(['United-States','Outlying-US(Guam-USVI-etc)'],'US')
census1['Native_country']=census1['Native_country'.replace('Scotland','UK')]
```

```python
#creating a function to find vif factor to check multi-collinearity
def cal_vif():
    viff=pd.DataFrame()
    viff["VIF FACTOR"]=[variance_inflation_factor(m.values,i) for i in range(m.shape[1])]
    viff["Features"]=m.columns
    print(viff)
```

# SPLITTING TRAIN and TEST DATA

```python
X = census1.drop(columns =['Income'], axis = 1)
Y = census1['Income']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=42, test_size=0.3)
```

# STANDARDIZING DATA

```python
#normalising data, it wll normalise data such that mean =1, std=0
sc = StandardScaler()
#standard scaler is done only on x train and xtest, it is generally not done on y train and y test.
#fitting the data into training part of x and y.
X_tr = sc.fit_transform(X_train)
X_tt = sc.transform(X_test)
```

# MODEL BUILDING - CLASSIFICATION

# 1.DECISION TREE CLASSIFIER

```python
dtr = DecisionTreeClassifier() #creating an instance
dtr.fit(X_tr,Y_train)  #fitting random forest model into trained data
ytr_pred=dtr.predict(X_tt)  #predicting testing data
print('accuracy=', metrics.accuracy_score(Y_test,ytr_pred))
print('cv score=',cross_val_score(dtr, X, Y, cv=5))
#accuracy of this model is 81%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of income <=50k, income>50k of actual and predicted y value
comat1 = confusion_matrix(Y_test,ytr_pred)
col = ['INCOME >50K','INCOME<=50K']
row = ['Predicted INCOME<=50K','Predicted INCOME >50K']
comat_frame = pd.DataFrame(comat1,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscoreTRee = dtr.predict_proba(X_tt)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(Y_test, yscoreTRee)
print('roc_auc_score for decision tree classifier is ', roc_auc_score(Y_test, yscoreTRee))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('decision tree - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# 2. RANDOM FOREST CLASSIFIER

```python
RF = RandomForestClassifier(n_estimators=100)#creating an instance
RF.fit(X_tr,Y_train)  #fitting random forest model into trained data
yRF_pred=RF.predict(X_tt) #predicting testing data
print('accuracy=', metrics.accuracy_score(Y_test,yRF_pred))
print('cv score=',cross_val_score(RF, X, Y, cv=5))
# our models is working well for 85.3%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of income <=50k, income>50k of actual and predicted y value
comat2 = confusion_matrix(Y_test,yRF_pred)
col = ['INCOME >50K','INCOME<=50K']
row = ['Predicted INCOME<=50K','Predicted INCOME >50K']
comat_frame = pd.DataFrame(comat2,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
yscore_RF = RF.predict_proba(X_tt)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(Y_test, yscore_RF)
print('roc_auc_score for LOGISTIC REGRESSION classifier is ', roc_auc_score(Y_test, yscore_RF))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,5))
plt.title('RANDOM FOREST CLASSIFIER - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# HYPERPARAMETER TUNING

```python
#as random forest regressor is the best model and has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
RFR = RandomForestClassifier() #creating an instance
n_estimators = [0.1,10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grid = dict(n_estimators = n_estimators)
gridforest = GridSearchCV(estimator = RFR, param_grid = grid, n_jobs=-1, verbose = 2, cv = 5)
gridforest.fit(X_tr,Y_train) #fitting the model into trained data
```
