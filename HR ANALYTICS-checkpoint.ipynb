# SPLITTING OF DATA

```python
#partitioning
#splitting the train data into train and test
#dividing data>>>
x = hr1.drop(['Attrition'], axis = 1)
y = hr1['Attrition']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=47, test_size=0.25)
```

```python
#scaling of data and checking if it fits and transforming split train and test data
sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)
```

# model building

# RANDOM FOREST CLASSIFIER

```python
clf = RandomForestClassifier(n_estimators=100)#creating an instance
clf.fit(x_train_sc,y_train)  #fitting random forest model into trained data
```

```python
clf_pred=clf.predict(x_test_sc) #predicting testing data
```

```python
print('accuracy=', metrics.accuracy_score(y_test,clf_pred))
print('cv score=',cross_val_score(clf, x, y, cv=5))
# our models is working well for 84%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat = confusion_matrix(y_test,clf_pred)
col = ['In the company','left the company']
row = ['Predicted in the company','Predicted left the company']
comat_frame = pd.DataFrame(comat,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
y1score = clf.predict_proba(x_test_sc)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y1score)
print('roc_auc_score for random forest classifier is ', roc_auc_score(y_test, y1score))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,10))
plt.title('random forest - Receiver Operating Characteristics')
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

# HYPERPARAMETER TUNING

```python
#as logistic regression has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
clr = LogisticRegression()
solvers = ['newton-cg','lbfgs','liblinear']
penalty = ['l1','l2']
c_values = [100,10,1.0,0.1,0.01,0.001,0.0001]
```

```python
grid = dict(solver = solvers, penalty = penalty, C=c_values)
cv = RepeatedStratifiedKFold(n_splits =10,n_repeats = 5,random_state=42)
```

```python
grid_s = GridSearchCV(estimator = clr, param_grid = grid, n_jobs=-1, verbose = True, cv = cv, scoring = 'accuracy', error_score = 0)
grid_s.fit(x_train_sc,y_train) #fitting into trained data
```

```python
grid_s.best_params_ #selecting best parameter
```

```python
grid_s.best_score_ #getting best score
```

```python
gridpred = grid_s.predict(x_test_sc) #predicting test data 
gridpred
```

```python
print('accuracy=', metrics.accuracy_score(y_test,gridpred))
print('cv score=',cross_val_score(clr, x, y, cv=5))
# our models is working well for 85%
```

```python
#creating the confusion matrix which is a graphical 
#representation of data of yes - attrition( left the company), no-attrition(in the company) of actual and predicted y value
comat = confusion_matrix(y_test,gridpred)
col = ['In the company','left the company']
row = ['Predicted in the company','Predicted left the company']
comat_frame = pd.DataFrame(comat,index = row, columns = col)
plt.figure.figsize = (20,10)
plt.subplot()
sn.heatmap(comat_frame, annot = True)
```

```python
#roc auc score calculation
y4score = grid_s.predict_proba(x_test_sc)[:,1]
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_test, y4score)
print('roc_auc_score after hypertuning is ', roc_auc_score(y_test, y4score))
```

```python
#graphical representation of roc auc curve
plt.subplots(1, figsize=(10,10))
plt.plot(false_positive_rate1, true_positive_rate1)
plt.plot([0, 1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7"), plt.plot([1, 1] , c=".7")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
