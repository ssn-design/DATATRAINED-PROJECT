```python
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df.head()
df.rename(columns = {'class':'type'}, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
X_df = df.drop(columns = 'type')
```

```python
Y_df = df['type']
```

```python
#Label Encoding

for col in X_df.columns:
    X_df[col] = le.fit_transform(X_df[col])

le_ = LabelEncoder()
Y_df = le_.fit_transform(Y_df)
```

```python
#One hot encoding for categorical data

X_df = pd.get_dummies(X_df, columns = X_df.columns, drop_first = True)
```

```python
#Train-test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, Y_df, test_size = 0.2)
```

```python
#Using logistic regression

from sklearn.linear_model import LogisticRegression

L_reg = LogisticRegression()
L_reg.fit(X_train, y_train)
```

```python
#Checking the R2-score of our regression model

L_reg.score(X_test, y_test)
```

```python
#Predictions of our LR model

predicted = L_reg.predict(X_df)
```

```python
#Decoding back to categorical variables

predicted_class = le_.inverse_transform(predicted)
```

```python
#Inserting predicted type of mushroom in the original dataframe

df.insert(loc = 1, column = 'Predicted type', value = predicted_class)
```
