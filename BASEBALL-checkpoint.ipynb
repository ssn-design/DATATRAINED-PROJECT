# feature enginnering

```python
#feature selection. dropping features not required.
baseball.drop(['RUNS ALLOWED','EARNED RUNS','EARNED RUNS AVG'],axis = 1, inplace = True)
```

```python
#shape of the dataset after dropping of columns.
#we have 14 columns now from 17
baseball.shape
```

# splitting data

```python
#label column is wins.we are separating the data into features=x and label=y 
x = baseball.drop('WINS', axis = 1)
y = baseball['WINS']
```

# training data

```python
#training the data 
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

# standardizing

```python
#standardization
#normalising data, it wll normalise data such that mean =1, std=0
sc = StandardScaler()
#standard scaler is done only on x train and xtest, it is generally not done on y train and y test.
```

# fitting of x and y

```python
#fitting the data into training part of x and y.
x_tr = sc.fit_transform(x_train)
x_te = sc.transform(x_test)
```

# model building

# linear regression- model1

```python
lr = LinearRegression() #creating an instance
lr.fit(x_tr,y_train) #fitting linear regression model into trained data
lr.coef_ #coefficient of x
```

```python
lr.intercept_  #intercept value of the regreesion model
```

```python
lr.score(x_tr,y_train) #as per the score, our model works 94% well
```

```python
#predict the value
ypred_lr = lr.predict(x_te)
print("predicted result wins:",ypred_lr)
print('actual wins:',y_test)
print('r2 score =',r2_score(y_test,ypred_lr))
print('mean absolute error =',mean_absolute_error(y_test,ypred_lr))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,ypred_lr)))
print('cv score=',cross_val_score(lr, x, y, cv=3))
```

# ElasticNet regression

```python
ent = ElasticNet(alpha = 0.0001) #creating an instance
ent.fit(x_tr,y_train)   #fitting into trained data
ent.score(x_tr,y_train)  #as per the score, our model works 94% well
```

```python
#predict the value, r2 and error scores
predy_ent = ent.predict(x_te)
print("predicted result wins:",predy_ent)
print('actual wins:',y_test)
print('r2 score =',r2_score(y_test,predy_ent))
print('mean absolute error =',mean_absolute_error(y_test,predy_ent))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predy_ent)))
print('cv score=',cross_val_score(ent, x, y, cv=3))
```

# RANDOM FOREST REGRESSOR

```python
rf = RandomForestRegressor() #creating an instance
rf.fit(x_tr,y_train) #fitting random forest regressor into x train and y train values
```

```python
rf.score(x_tr,y_train)  #as per the score, our model works 91% well
```

```python
#predict the value, r2 and error scores
predy_rf = rf.predict(x_te)
print("predicted result wins:",predy_rf)
print('actual wins:',y_test)
print('r2 score =',r2_score(y_test,predy_rf))
print('mean absolute error =',mean_absolute_error(y_test,predy_rf))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predy_rf)))
print('cv score=',cross_val_score(rf, x, y, cv=3))
```

# KNNEIGHBORS REGRESSOR

```python
knn = KNeighborsRegressor(n_neighbors=7) #creating an instance
KNeighborsRegressor(algorithm='auto', leaf_size=30, metric='minkowski',
          metric_params=None, n_jobs=1, n_neighbors=7, p=2,
          weights='uniform')
```

```python
knn.fit(x_tr,y_train) #fitting kneighbors into x train and y train values
```

```python
knn.score(x_tr,y_train)  #as per the score, our model works 55% well
```

```python
#predict the value, r2 and error scores
predy_knn = knn.predict(x_te)
print("predicted result wins:",predy_knn)
print('actual wins:',y_test)
print('r2 score =',r2_score(y_test,predy_knn))
print('mean absolute error =',mean_absolute_error(y_test,predy_knn))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,predy_knn)))
print('cv score=',cross_val_score(knn, x, y, cv=3))
```

# HYPERPARAMETER TUNING

```python
#Elasticnet being the best model, we are proceeding with its tuning
parametergrid = {'max_iter':[1,5,10],'alpha':[0.0001,0.001,0.01,0.1,1,10,100],'l1_ratio':np.arange(0.0,1.0,0.1)}
ENET = ElasticNet() #creating an instance
grd = GridSearchCV(ENET,parametergrid, scoring ='r2', cv = 10)
grd.fit(x_tr,y_train)
grid_pred_y = grd.predict(x_te)
grid_pred_y
```

```python
grd.score(x_tr,y_train) #as per the score, our model works 83% well
```

```python
#predict the value, r2 and error scores
print("predicted result wins:",grid_pred_y)
print('actual wins:',y_test)
print('r2 score =',r2_score(y_test,grid_pred_y))
print('mean absolute error =',mean_absolute_error(y_test,grid_pred_y))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,grid_pred_y)))
print('cv score=',cross_val_score(ENET, x, y, cv=5))
```
