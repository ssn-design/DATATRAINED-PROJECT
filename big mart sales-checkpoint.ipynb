# handling missing values

```python
#checking mean value of item weight. missing values will be replaced by mean(train)
dt_tr['Item_Weight'].mean()
```

```python
#checking mode value of outlet size. missing values will be replaced by mode(train)
dt_tr['Outlet_Size'].mode()
```

```python
#checking mean value of item weight. missing values will be replaced by mean(test)
dt_te['Item_Weight'].mean()
```

```python
#checking mode value of outlet size. missing values will be replaced by mode(test)
dt_te['Outlet_Size'].mode()
```

```python
#filling missing values. (both train & test dataset)
#as item weight is numerical, so we used mean imputation i'e filled the null values with mean value.
#outlet size is categorical. so we used mode imputation i'e filled the null values with mode value.
dt_tr['Item_Weight'].fillna(dt_tr['Item_Weight'].mean(), inplace = True)
dt_tr['Outlet_Size'].fillna(dt_tr['Outlet_Size'].mode()[0],inplace = True)
dt_te['Item_Weight'].fillna(dt_te['Item_Weight'].mean(), inplace = True)
dt_te['Outlet_Size'].fillna(dt_te['Outlet_Size'].mode()[0],inplace = True)
```

```python
#assigning numerical values to the categorical values of columns in the dictionary-train
new_data_train = {'Item_Fat_Content':{'Low Fat':1,'Regular':2,'LF':3,'reg':4,'low fat':1},'Item_Type':{'Fruits and Vegetables':1,'Snack Foods':2,'Household':3,'Frozen Foods':4,"Dairy":5,'Canned':6,'Baking Goods':7,'Health and Hygiene':8,'Soft Drinks':9,'Meat':10,'Breads':11,'Hard Drinks':12,'Others':13,'Starchy Foods':14,'Breakfast':15,'Seafood':16},'Outlet_Size':{'Medium':2,'Small':1,'High':3},'Outlet_Location_Type':{'Tier 3':3,'Tier 2':2,'Tier 1':1},'Outlet_Type':{'Supermarket Type1':1,'Supermarket Type2':2,'Supermarket Type3':3,'Grocery Store':4}}
```

```python
#replacing categorical data with assigned numerical data
dt_tr1 = dt_tr.replace(new_data_train)
dt_tr1.head()
```

```python
#assigning numerical values to the categorical values of columns in the dictionary - test
new_data_test = {'Item_Fat_Content':{'Low Fat':1,'Regular':2,'LF':3,'reg':4,'low fat':1},'Item_Type':{'Fruits and Vegetables':1,'Snack Foods':2,'Household':3,'Frozen Foods':4,"Dairy":5,'Canned':6,'Baking Goods':7,'Health and Hygiene':8,'Soft Drinks':9,'Meat':10,'Breads':11,'Hard Drinks':12,'Others':13,'Starchy Foods':14,'Breakfast':15,'Seafood':16},'Outlet_Size':{'Medium':2,'Small':1,'High':3},'Outlet_Location_Type':{'Tier 3':3,'Tier 2':2,'Tier 1':1},'Outlet_Type':{'Supermarket Type1':1,'Supermarket Type2':2,'Supermarket Type3':3,'Grocery Store':4}}
```

```python
#replacing categorical data with assigned numerical data
dt_te1 = dt_te.replace(new_data_test)
dt_te1.head()
```

```python
#partitioning
#splitting the train data into train and test
#dividing data>>>
x = dt_tr1.drop('Item_Outlet_Sales', axis = 1)
y = dt_tr1['Item_Outlet_Sales']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=42, test_size=0.2)
```

```python
#standardization
#normalising data, it wll normalise data such that mean =1, std=0
#standard scaler is not required for random tree classifier and decision tree
x.describe()
```

```python
sc = StandardScaler()
#standard scaler is done only on x train and xtest, it is generally not done on y train and y test.
```

```python
#fit is applied only on xtrain as fit means learning how the data is distributed in x train and then transform it.
x_tr = sc.fit_transform(x_train)
x_tr
```

```python
#transforming x test by usung scaler and normalizing it
x_te = sc.transform(x_test)
x_te
```

# model building

# LINEAR REGRESSION

```python
LR = LinearRegression() 
```

```python
LR.fit(x_tr, y_train) #fitting linear regression into x train and y train values
```

```python
y_pred_LR = LR.predict(x_te) #predicting new y test
```

```python
#checking its performance
#model has 49% accuracy
print('r2 score =',r2_score(y_test,y_pred_LR))
print('mean absolute error =',mean_absolute_error(y_test,y_pred_LR))
print('mean squared error =',np.sqrt(mean_squared_error(y_test,y_pred_LR)))
```

# RANDOM FOREST REGRESSOR

```python
y_pred_RF = rf.predict(x_te) #predicting new y test
```
