# CHECK MULTI COLLINEARITY

```python
#separating target column
x1 = flt_train.drop(columns =['Price'], axis = 1, inplace=False)
```

```python
#creating a fuction to find vif factor to check multi-collinearity
def calvif():
    vif=pd.DataFrame()
    vif["VIF FACTOR"]=[variance_inflation_factor(x1.values,i) for i in range(x1.shape[1])]
    vif["Features"]=x1.columns
    print(vif)
```

# SPLITTING TRAIN DATA

```python
#splitting x and y to train
X=flt_train1.drop(['Price'],axis=1)
Y=flt_train1['Price']
```

```python
#training data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 65) 
```

# standardization

```python
#normalising data, it wll normalise data such that mean =1, std=0
sc = StandardScaler()
#standard scaler is done only on x train and xtest, it is generally not done on y train and y test.
#fitting the data into training part of x and y.
X_tr = sc.fit_transform(X_train)
X_tt = sc.transform(X_test)
```

# MODEL BUILDING : 1.LINEAR REGRESSION

```python
lr = LinearRegression() #creating an instance
lr.fit(X_tr,Y_train) #fitting linear regression model into trained data
lr.coef_ #coefficient of x
```

```python
lr.intercept_  #intercept value of the regreesion model
```

```python
lr.score(X_tr,Y_train) #as per the score, our model works 52% well
```

```python
#predict the value
ypred_lr = lr.predict(X_tt)
print("predicted flight price:",ypred_lr)
print('actual flight price:',Y_test)
print('r2 score =',r2_score(Y_test,ypred_lr))
print('mean absolute error =',mean_absolute_error(Y_test,ypred_lr))
print('mean squared error =',np.sqrt(mean_squared_error(Y_test,ypred_lr)))
print('cv score=',cross_val_score(lr, X, Y, cv=3))
```

# RANDOM FOREST REGRESSOR

```python
rfr = RandomForestRegressor() #creating an instance
rfr.fit(X_tr,Y_train) #fitting random forest regressor into x train and y train values
rfr.score(X_tr,Y_train)  #as per the score, our model works 93% well
```

```python
#predict the value, r2 and error scores
predy_rfr = rfr.predict(X_tt)
print("predicted result prices :",predy_rfr)
print('actual prices:',Y_test)
print('r2 score =',r2_score(Y_test,predy_rfr))
print('mean absolute error =',mean_absolute_error(Y_test,predy_rfr))
print('mean squared error =',np.sqrt(mean_squared_error(Y_test,predy_rfr)))
print('cv score=',cross_val_score(rfr, X, Y, cv=5))
```

# HYPERPARAMETER TUNING

```python
#as random forest regressor is the best model and has least difference between accuracy and cv score. thus we are proceeding with its tuning
#we are using grid search cv
RFR = RandomForestRegressor() #creating an instance
n_estimators = [0.1,10,100,1000]
maxdepth = range(1,110)
minsampleleaf = np.linspace(0.1,1.0)
mostfeatures = ['auto','sqrt','log2']
minsamplesplit= np.linspace(0.1,1.0, 10)
grid = dict(n_estimators = n_estimators)
gridforest = GridSearchCV(estimator = RFR, param_grid = grid, n_jobs=-1, verbose = 2, cv = 7)
gridforest.fit(X_tr,Y_train) #fitting the model into trained data
```
